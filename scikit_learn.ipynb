```python
#Return deviation--to determine risk and return 

#expected return measures the mean, or expected value, of the probability distribution of investment returns. 
#the expected return of a portfolio is calculated by multiplying the weight of each asset by its expected return...
#...and adding the values for each investment
returns = close_price / close_price.shift(1) - 1
returns.plot(label='return')
```

```python
start_date = '2013-01-01'
end_date = '2023-01-01'

symbols_list = ['AAPL','META','NVDA','TSLA','GOOGL','AMZN','IBM','KO','MSFT',"VTI"]

data_set_competitors = pd.DataFrame(columns=symbols_list)

for symbol in symbols_list:
    data_set_competitors[symbol] = yf.download(symbol, start_date, end_date)['Close']
    
data_set_competitors
```

```python
#analyzing AAPL against other tech giants
data_set_competitors.plot(figsize=(10,7))
plt.legend()
plt.title("Adjusted Close Price", fontsize=16)
plt.ylabel("Price ($)", fontsize=14)
plt.xlabel("Year", fontsize=14)
plt.grid(which='major', color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```python
#correlation analysis--Does one competitor affect another? 
#we can analyze the competition by percentage change and correlation function in pandas.
#percentage change will find how much the price changes compared to the previous day which defines returns
returnscompetition = data_set_competitors.pct_change()
correlation = returnscompetition.corr()
correlation #prints in matrix formation
```

```python
#because VTI is apple's highest share holder it makes sense its almost 70%. but for MSFT being highest competitor
#lets do a scatterplot of apple's return distributions vs microsoft
plt.scatter(returnscompetition.AAPL, returnscompetition.MSFT)
plt.xlabel('Returns from AAPL')
plt.ylabel("Returns from MSFT")
```

```python
#there is a pretty positive linear correlation, the higher apple the higher is msft
#we are going to run a Kernel Density Estimate of this, it is a fundamental data smoothing problem
#inferences about the pobulation are made, based on a finite data sample. it helps generate estimations of the overall
#distributions
pd.plotting.scatter_matrix(returnscompetition, diagonal='kde', figsize=(10,10));
```

```python
#In order to prove positive correlations, we will use heat maps to visualize the correlation ranges among comp stock
#lighter the color the more correlated the two stocks are.
plt.imshow(correlation, cmap='hot', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation)), correlation.columns)
plt.yticks(range(len(correlation)), correlation.columns);
```

```python
#from scatter matrix and heatmap, we can find great correlations among the competing stocks. 
#however this does not show causality, just that there is a trend in the tech space among competiting stocks
#rather than how competiting stocks affect each other

#A GOOD PLACE TO UPGRADE THIS WOULD BE HOW DO COMPETITING STOCKS AFFECT ONE ANOTHER
```

```python
#Predicting Stocks Price
#we're going to use Simple Linear Analysis, Quadratic Discriminant Analysis (QDA), and KNN
#we'll first add high low percentage and percentage change
df_regular_AAPL = apple_data_df.loc[:,['Adj Close', 'Volume']]
df_regular_AAPL["HL_PCT"] = (apple_data_df['High'] - apple_data_df['Low']) / apple_data_df['Close'] * 100.0
#high low percent change = high - low / close (total) * 100
df_regular_AAPL['PCT_Change'] = (apple_data_df["Close"] - apple_data_df['Open']) / apple_data_df['Open'] * 100.0
```

```python
df_regular_AAPL #yearly by month
```

```python
import numpy as np
import math
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import preprocessing, svm
#preprocessing & Cross Validation
#first drop missing values
df_regular_AAPL.fillna(value=-99999, inplace=True)
print(df_regular_AAPL.shape)

#we want to separate 1 percent of the data to forecast
forecast_out = int(math.ceil(0.01*len(df_regular_AAPL)))

#separating the label, we want to predict adj close
forecast_close = 'Adj Close'
df_regular_AAPL['label'] = df_regular_AAPL[forecast_close].shift(-forecast_out)
X = np.array(df_regular_AAPL.drop(['label'],1))

#scale the X so that everyone can have the same distribution for linear regression
X = preprocessing.scale(X)

#find data series of late X and early X by training for model generation and evaluation
X_lately = X[-forecast_out:]
X = X[:-forecast_out]

#separate label and identify it as target y
y = np.array(df_regular_AAPL['label'])
y = y[:-forecast_out]

print('Dimension of X', X.shape)
print('Dimension of y', y.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
#separation of training and testing of model by cross validation train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clfreg = LinearRegression(n_jobs=-1)
clfreg.fit(X_train, y_train)

clfpoly2 = make_pipeline(PolynomialFeatures(2), Ridge())
clfpoly2.fit(X_train, y_train)

clfpoly3 = make_pipeline(PolynomialFeatures(3), Ridge())
clfpoly3.fit(X_train, y_train)
```

```python
#KNN regression
clfknn = KNeighborsRegressor(n_neighbors=2)
clfknn.fit(X_train, y_train)
```

```python
#evaluation
confidence_reg = clfreg.score(X_test, y_test)
confidencepoly2 = clfpoly2.score(X_test, y_test)
confidencepoly3 = clfpoly3.score(X_test, y_test)
confidenceknn = clfknn.score(X_test, y_test)

#results
print("The linear regression confidence is ", confidence_reg)
print("The quadratic regression 2 confidence is ", confidencepoly2)
print("The quadratic regression 3 confidence is ", confidencepoly3)
print("The KNN regression confidence is ", confidenceknn)
```

```python
forecast_set_knn = clfknn.predict(X_lately)
df_regular_AAPL['Forecast'] = np.nan
print(forecast_set_knn, confidenceknn, forecast_out)
```

```python
#plotting predictions
last_date = df_regular_AAPL.iloc[-1].name
last_unix = last_date
next_unix = last_unix + datetime.timedelta(days=1)

for i in forecast_set_knn:
    next_date = next_unix
    next_unix += datetime.timedelta(days=1)
    df_regular_AAPL.loc[next_date] = [np.nan for _ in range(len(df_regular_AAPL.columns)-1)]+[i]
```

```python
df_regular_AAPL['Adj Close'].tail(500).plot()
df_regular_AAPL['Forecast'].tail(500).plot()
plt.legend(loc=4)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.show()
```
